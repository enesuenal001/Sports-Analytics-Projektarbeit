```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data = pd.read_csv('/Users/Enes/Documents/Daten_Projektarbeit/data_final.csv', encoding="UTF-8")
data.head()
```

```python
#Trennung zwischen nicht zu normalisierenden und normalisierenden Spalten
no_norm = ['Player', 'Pos', 'Squad', 'Age', '90s', 'Transfered']
norm = [x for x in list(data) if x not in no_norm]
```

```python
data[['Gls']].plot()
data[['G-xG']].plot()
data[['Touches']].plot()
plt.savefig('before_norm.png')
```

```python
#Normalisieren der numerischen Datenstze von der Spalte 'Goals' bis 'Won%'
values = data[norm].values
scaler = MinMaxScaler()
values_norm = scaler.fit_transform(values)
data_norm = pd.DataFrame(values_norm, columns = norm, index = data.index)
data[norm] = data_norm
data
```

```python
data[['Gls']].plot()
data[['G-xG']].plot()
data[['Touches']].plot()
plt.savefig('after_norm.png')
```

```python
data.to_csv('data_norm.csv', encoding="UTF-8", index=False)
```

```python
data[data['Pos'] == 'FW'].to_csv('norm_fw.csv', encoding="UTF-8", index=False)
data[data['Pos'] == 'MF'].to_csv('norm_mf.csv', encoding="UTF-8", index=False)
data[data['Pos'] == 'DF'].to_csv('norm_df.csv', encoding="UTF-8", index=False)
```

```python
pca = PCA()
data_pca = pca.fit_transform(data_norm)
pca.explained_variance_ratio_
np.cumsum(pca.explained_variance_ratio_ * 100)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Erluterte Varianz in % (Informationsgehalt)')
plt.grid()
plt.savefig('pca.png', dpi=100)
```

```python
print('Erluterte Varianz in % bei einem Hauptkomponenten: ',
np.cumsum(pca.explained_variance_ratio_ * 100)[0])

print('Erluterte Varianz in % bei zwei Hauptkomponenten: ',
np.cumsum(pca.explained_variance_ratio_ * 100)[1])

print('Erluterte Varianz in % bei drei Hauptkomponenten: ',
np.cumsum(pca.explained_variance_ratio_ * 100)[2])
```

```python
comp = pd.DataFrame(data_pca[:,:2])
comp
```

```python
data = pd.concat([data, comp], axis=1)
```

```python
data.rename(columns={0: 'PC1', 1: 'PC2'}, inplace=True)
data
```

```python
data[data['Pos'] == 'FW'].to_csv('pca_fw.csv', encoding="UTF-8", index=False)
data[data['Pos'] == 'MF'].to_csv('pca_mf.csv', encoding="UTF-8", index=False)
data[data['Pos'] == 'DF'].to_csv('pca_df.csv', encoding="UTF-8", index=False)
```
