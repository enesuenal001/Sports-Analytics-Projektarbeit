```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data = pd.read_csv('/Users/Enes/Documents/Daten_Projektarbeit/data_final.csv', encoding="ISO-8859-1")
data.head()
```

```python
#Trennung zwischen nicht zu normalisierenden und normalisierenden Spalten
no_norm = ['Player', 'Pos', 'Squad', 'Age', '90s', 'Transfered']
norm = [x for x in list(data) if x not in no_norm]
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
#Normalisieren der numerischen Datensätze von der Spalte 'Goals' bis 'Won%'
values = data[norm].values
scaler = StandardScaler()
values_norm = scaler.fit_transform(values)
data_norm = pd.DataFrame(values_norm, columns = norm, index = data.index)
data[norm] = data_norm
data
```

```python
data[['Gls']].plot()
data[['G-xG']].plot()
data[['Rec']].plot()
```
