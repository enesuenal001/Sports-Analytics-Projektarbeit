```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data = pd.read_csv('/Users/Enes/Documents/Daten_Projektarbeit/Player_Data.csv', encoding="UTF-8")
data.head()
```

```python
data['Pos'] = data['Pos'].str[:2]
data['Player'] = data['Player'].str.split('\\').str[0]

data.fillna(0)

data.head()
```

```python
data_90s = data[data['90s'] > 7.0]
data = data_90s
data.head()
```

```python
data_field = data[data['Pos'] != 'GK']
data = data_field
data.head()
```

```python
data_transfers = data[data['Transfered'] == True]
data_transfers.head()
```

```python
data_normal = data[(data['Transfered'] == False) & (data['90s'] >= 10.0)].copy()
data_normal.head()
```

```python
data_final = pd.concat([data_normal, data_transfers], axis=0)
data_final.head()
```

```python
data_final.sort_values('Rk')
```

```python
del data_final['Rk']
data_final.head()
```

```python
data_final.reset_index(drop=True, inplace=True)
```

```python
data_final.sort_values(by='Player', key=lambda x: x.str.split('\s+').str[-1])
```

```python
data_final.reset_index()
```

```python
data_final = data_final.sort_values(by='Player', key=lambda x: x.str.split('\s+').str[-1])
data_final
```

```python
#Der Spieler Kevin Vogt war  mit Daten aus zwei Vereinen in der Tabelle aufzufinden; die Daten aus der Hinrunde werden entfernt
data_final = data_final.drop(data_final[(data_final["Player"] == "Kevin Vogt") & (data_final["Squad"] == "Hoffenheim")].index)
data_final.reset_index(drop=True)
```

```python
data_final.to_csv('data_final.csv', encoding="UTF-8", index=False)
```
